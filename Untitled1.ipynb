```python
f = open("output.txt", "r+")

episodeNo = []
rewardPerEpisode = []
maxRewardByEpisode = []
epsilonDecayByEpisode = []

for line in f.readlines():
    listOfWords = line.split()
    episodeNo.append(int(listOfWords[1]))
    rewardPerEpisode.append(int(float(listOfWords[4])))
    maxRewardByEpisode.append(int(float(listOfWords[7])))
    epsilonDecayByEpisode.append(float(listOfWords[9]))
    print (listOfWords[9])
```

```python
from matplotlib.pyplot import plot
```

```python
plot(episodeNo, rewardPerEpisode, color='brown', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plot(episodeNo, maxRewardByEpisode, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plot(episodeNo, epsilonDecayByEpisode, color='grey', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```
